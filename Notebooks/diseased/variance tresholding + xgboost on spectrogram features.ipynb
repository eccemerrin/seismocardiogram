```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
```

## Load data

```python
spec = pd.read_csv(csv_path + "/spectogram_coef.csv", index_col = 0)
spec
```

Here, our target is to predict disease correctly. So; our y will be disease column of spec dataset

```python
spec_n = spec.drop(['disease'], axis =1)
y = spec.disease
```

```python
y
```

```python
y_dummy = pd.get_dummies(y)
y_dummy
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(spec_n, y_dummy, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
y_valid
```

```python
from xgboost import XGBRegressor

# Define the model
my_model_1 = XGBRegressor(random_state = 0) # Your code here

# Fit the model
my_model_1.fit(X_train, y_train)# Your code here
```

```python
predictions_1 = my_model_1.predict(X_valid)
```

```python
from sklearn.metrics import mean_absolute_error

mae_1 = mean_absolute_error(predictions_1,y_valid) # Your code here

# Uncomment to print MAE
print("Mean Absolute Error:" , mae_1)
```

## Varience Thresholding

### with mean variation

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
spec_n.var()
```

```python
spec_n.var().mean()
```

```python
vt = VarianceThreshold(threshold = spec_n.var().mean())
```

```python
_ = vt.fit(spec_n)
mask_spec = vt.get_support()
t_spec= spec_n.loc[:, mask_spec]
t_spec
```

```python
X_train_t, X_valid_t, y_train_t, y_valid_t = train_test_split(t_spec, y_dummy, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
model = XGBRegressor(random_state = 0) # Your code here

# Fit the model
model.fit(X_train_t, y_train_t)# Your code here

```

```python
predictions_t = model.predict(X_valid_t)
```

```python
mae_t = mean_absolute_error(predictions_t,y_valid_t) # Your code here

# Uncomment to print MAE
print("Mean Absolute Error:" , mae_t)
```

### with 1/10th of mean variation

```python
vt_10 = VarianceThreshold(threshold = spec_n.var().mean()/10)
```

```python
_ = vt_10.fit(spec_n)
mask_spec = vt_10.get_support()
t_10_spec= spec_n.loc[:, mask_spec]
t_10_spec
```

```python
X_train_t10, X_valid_t10, y_train_t10, y_valid_t10 = train_test_split(t_10_spec, y_dummy, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
X_valid_t10.shape
```

```python
model_10  = XGBRegressor(random_state = 0) # Your code here

# Fit the model
model_10.fit(X_train_t10, y_train_t10)# Your code here
```

```python
predictions_t10 = model_10.predict(X_valid_t10)
```

```python
mae_t10 = mean_absolute_error(predictions_t10, y_valid_t10) # Your code here

# Uncomment to print MAE
print("Mean Absolute Error:" , mae_t)
```

```python
n_estimators = list(np.arange(100,1000,100))
scores = []
for i in n_estimators:
    
    my_model = XGBRegressor(n_estimators = i, learning_rate =0.2)
    my_model.fit(X_train_t10, y_train_t10, early_stopping_rounds = 5, eval_set =[(X_valid_t10, y_valid_t10)],verbose = False)
    pred = my_model.predict(X_valid_t10)
    mae = mean_absolute_error(pred, y_valid_t10) 
    scores.append(mae)

print(dict(zip(n_estimators, scores)))
```
