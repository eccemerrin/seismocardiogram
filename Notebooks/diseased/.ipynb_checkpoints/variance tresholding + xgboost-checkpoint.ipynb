```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vt = VarianceThreshold(threshold = 0.0004)
```

```python
_ = vt.fit(spec_num)
mask_spec = vt.get_support()
new_spec = spec_num.loc[:, mask_spec]
new_spec
```

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
categorical_pipeline = Pipeline(
    steps=[
        ("impute", SimpleImputer(strategy="most_frequent")),
        ("oh-encode", OneHotEncoder(handle_unknown="ignore", sparse=False)),
    ]
)
```

```python

numeric_pipeline = Pipeline(
    steps=[("impute", SimpleImputer(strategy="mean")), 
           ("scale", StandardScaler())]
)
```

```python
cat_cols = spec.select_dtypes(exclude="number").columns
num_cols = new_spec.select_dtypes(include="number").columns
```

```python
from sklearn.compose import ColumnTransformer

full_processor = ColumnTransformer(
    transformers=[
        ("numeric", numeric_pipeline, num_cols),
        ("categorical", categorical_pipeline, cat_cols),
    ]
)
```

```python
import xgboost as xgb

xgb_cl = xgb.XGBClassifier()
```

```python
X_processed = full_processor.fit_transform(spec)
y_processed = SimpleImputer(strategy="most_frequent").fit_transform(
    y.values.reshape(-1, 1)
)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y_processed, stratify=y_processed, random_state=1121218
)
```
