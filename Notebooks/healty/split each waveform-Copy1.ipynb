```python
def splitter(array, slicing_point = 10000, duration = 10):
    
    wanted_npts = int(array[0].stats.sampling_rate * duration) #2560
    npts_x = int(array[0].stats.npts - slicing_point)
    npts_y = int(array[1].stats.npts - slicing_point)
    npts_z = int(array[2].stats.npts - slicing_point)
    
    lenght_x = npts_x // wanted_npts
    lenght_y = npts_y // wanted_npts
    lenght_z = npts_z // wanted_npts
    
    remainder_x = int(np.around((((npts_x / wanted_npts - lenght_x)) * wanted_npts), 0))
    remainder_y = int(np.around((((npts_y / wanted_npts - lenght_y)) * wanted_npts), 0))
    remainder_z = int(np.around((((npts_z / wanted_npts - lenght_z)) * wanted_npts), 0))


    
    return [np.array_split(array[0][slicing_point + remainder_x:],lenght_x),
           np.array_split(array[1][slicing_point + remainder_y:],lenght_y),
           np.array_split(array[2][slicing_point + remainder_z:],lenght_z)]
    
```

```python
def merger(array):
    _x, _y, _z  = [], [], []
    for i in range(len(array)):
        a, b, c = splitter(array[i])
        _x = _x + a
        _y = _y + b
        _z = _z + c
        
    return [_x,_y,_z]
```
